### Filtering Data

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head(5)
```

```python
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
df.head(10)
```

```python
cols_set1 = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']
cols_set2 = ['math score', 'reading score', 'writing score']
df = df[cols_set1 + ['total score'] + cols_set2]

df
```

```python
#filter using specific data in csv file 
df.loc[df['parental level of education'] == "bachelor's degree"]
```

```python
#filter using multiple data in csv file
df.loc[(df['parental level of education'] == "bachelor's degree") & (df['test preparation course'] == "completed")]
```

```python
#specify limits for data in csv file 
new_df = df.loc[(df['parental level of education'] == "bachelor's degree") & (df['test preparation course'] == "completed") & (df['total score'] >= 220)]
new_df
```

```python
#reset index in csv file
new_df = new_df.reset_index()
new_df
```

```python
new_df = new_df.drop(['level_0'], axis=1)
new_df
```

```python
#filter using substring in csv file
new_df = new_df.loc[new_df['lunch'].str.contains('reduced')]
new_df
```

```python
#use regex to further filter in csv file
import re
new_df = new_df.loc[new_df['race/ethnicity'].str.contains("C|D", flags = re.I, regex=True)]
new_df
```

### Making Conditional Changes

```python
#change specific entry in csv file
df.loc[df["parental level of education"] == "bachelor's degree", "parental level of education"] = "Bachelor's"
df
```

```python
#using condition to change values in csv file
df.loc[df['total score'] > 220, ['test preparation course']] = 'GOOD WORK'
df
```

```python
#using condition to change values across multiple columns in csv file
df.loc[df['total score'] > 220, ['test preparation course', 'lunch']] = ['GOOD WORK', 'SPECIAL']
df
```

### Statistics

```python
#find mean data across unique column values in csv file
df.groupby(['gender']).mean().sort_values('total score', ascending=True)
```

```python
#find mean data across unique column values in csv file 
df.groupby(['race/ethnicity']).mean().sort_values('total score', ascending=False)
```

```python
#count specific values in columns in csv file
df['count'] = 1
df.groupby(['race/ethnicity']).count()['count']
```
